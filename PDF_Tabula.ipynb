```python
import tabula
import pandas as pd
```

```python
df = tabula.read_pdf('MGLU_ER_3T20_POR.pdf', pages = 5)
# o df é uma lista
tabela = df[0]
tabela = tabela.dropna(how='all',axis =0)
tabela = tabela.dropna(how='all',axis = 1)
tabela.columns = tabela.iloc[0]
tabela = tabela.reset_index(drop=True)
display(tabela)

#tabela.to_excel('tabela_dre.xlsx')
```

### Paginas com mais de uma tabela

```python
df1 = tabula.read_pdf('MGLU_ER_3T20_POR.pdf', pages = 12)
for tabela1 in df1:
  tabela1 = tabela1.dropna(how='all',axis =0)
  tabela1 = tabela1.dropna(how='all',axis = 1)
  display(tabela1)
```
