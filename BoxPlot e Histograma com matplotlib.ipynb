```python
# Importando a base
from sklearn.datasets import fetch_california_housing
```

```python
# Base importada
retorno = fetch_california_housing()
retorno.DESCR
```

```python
# DataFrame
import pandas as pd
casas = pd.DataFrame(retorno.data)
casas.columns = retorno.feature_names
casas['MedHouseVal'] = retorno.target
casas.head()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,2, sharex=True, figsize=(18,7))

ax[0,0].boxplot(casas['MedHouseVal'],patch_artist=True, medianprops={"color": "white", "linewidth": 0.5},boxprops={"facecolor": "#52688F", "edgecolor": "white","linewidth": 0.5},vert=False, labels=['MedHouseVal'],showfliers=True)
ax[0, 1].boxplot(casas['MedInc'],patch_artist=True, medianprops={"color": "white", "linewidth": 0.5},boxprops={"facecolor": "#7391C8", "edgecolor": "white","linewidth": 0.5},vert=False, labels=['MedInc'],showfliers=True)

ax[1, 0].hist(casas['MedHouseVal'],bins=20,color = '#52688F')
ax[1, 1].hist(casas['MedInc'],bins=20,color = '#7391C8')

ax[0,0].spines['right'].set_visible(False)
ax[0,0].spines['top'].set_visible(False)
ax[0,0].spines['left'].set_visible(False)
ax[0,0].spines['bottom'].set_visible(False)
ax[0,0].axis('off')
ax[0,0].set_title('MedHouseVal', color='Grey', size=24, loc='left')

ax[0,1].spines['right'].set_visible(False)
ax[0,1].spines['top'].set_visible(False)
ax[0,1].set_title('MedInc', color='Grey', size=24, loc='left')
ax[0,1].spines['left'].set_visible(False)
ax[0,1].spines['bottom'].set_visible(False)
ax[0,1].axis('off')

ax[1, 0].spines['right'].set_visible(False)
ax[1, 0].spines['top'].set_visible(False)


ax[1, 1].spines['right'].set_visible(False)
ax[1, 1].spines['top'].set_visible(False)


plt.show()
```
