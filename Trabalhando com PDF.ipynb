# Extraindo dados em PDF

- Pypdf ler pdf


```python
# trabalhar no pdf
import PyPDF2 as pyf
# manipular pastas
from pathlib import Path

# comandos mais usados para os PDF
# pyf.PdfFileReader
# pyf.PdfFileWriter
# pyf.PdfFileMerger
```

```python
nome_arquivo = "arquivo_ps.pdf"
```

```python
# lendo arquivo pdf
arquivo_pdf = pyf.PdfFileReader(nome_arquivo)
```

## Percorrendo as paginas e as separando

```python
i=1
for pagina in arquivo_pdf.pages:
    arquivo_novo = pyf.PdfFileWriter()
    arquivo_novo.addPage(pagina)
    #salvar o arquivo
    with Path(f'paginas/arquivo{i}.pdf').open(mode='wb') as arquivo_final:
        arquivo_novo.write(arquivo_final)
        i+=1
```

## Selecionando e juntando as paginas

```python
paginas = [1,2]

novo_pdf = pyf.PdfFileWriter()
for num_pagina in paginas:
    nome_arquivo = f'paginas/arquivo{num_pagina}.pdf'
    arquivo_pdf = pyf.PdfFileReader(nome_arquivo)
    pagina = arquivo_pdf.pages[0]
    novo_pdf.addPage(pagina)
    #pegar arquivo indicado no numero da pagina
    #adicionar no novo pdf

with Path(f'paginas/Consolidado.pdf').open(mode='wb') as arquivo_final:
    novo_pdf.write(arquivo_final)
```

## Mesclar arquivos

```python
pdf_mesclado = pyf.PdfFileMerger()

arquivo1 = 'paginas/arquivo1.pdf'
arquivo2 = 'paginas/arquivo2.pdf'

pdf_mesclado.append(arquivo1)
pdf_mesclado.append(arquivo2)

with Path(f'paginas/ConsolidadoAppend.pdf').open(mode='wb') as arquivo_final:
    pdf_mesclado.write(arquivo_final)
```

- Rodar pagina

```python
pdf_rodar = pyf.PdfFileReader(arquivo1)

pdf_final = pyf.PdfFileWriter()
for pagina in pdf_rodar.pages:
    pagina.rotateClockwise(90)
    pdf_final.addPage(pagina)
    
with Path(f'Arquivo_rotacionado.pdf').open(mode='wb') as arquivo_final:
    pdf_final.write(arquivo_final)
```

## Trabalhando com textos e informações dentro do PDF
- Pegar texto da pagina e Identificar onde está

```python
nome_arquivo_maga = 'MGLU_ER_4T20_POR.pdf'
arquivo_maga = pyf.PdfFileReader(nome_arquivo_maga)

#numero das paginas do arquivo
num_paginas = arquivo_maga.numPages
#informações dos arquivos
infos = arquivo_maga.getDocumentInfo()
```

```python
print(num_paginas)

print(infos)
```

```python
texto  = '''| 
Despesas com Vendas '''

i=1
for pagina in arquivo_maga.pages:
    #pegar o que está escrito na pagina
    texto_pag = pagina.extractText()
    #Verficar se existe o texto e se tiver pegamos o numero da pagina
    if texto in texto_pag:
        print(f'Está na pagina{i}')
        num_pagina = i
        texto_f = texto_pag
    i+=1
```

```python
print(texto_f)
```

```python
texto_f = texto_f.replace('  ',' ')
texto_f = texto_f.replace('\n \n','fffff')
texto_f = texto_f.replace('\n','')
texto_f = texto_f.replace('fffff','\n')
```
